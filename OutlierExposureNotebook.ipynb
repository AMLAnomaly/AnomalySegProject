```python
#Code to fine-tune EOMT with Outlier Exposure on Cityscapes

# specify in --data.init_args.path the path to the Cityscapes dataset
# specify in --data.init_args.coco_root the path to the COCO images already processed (one object for image + related mask)
# specify in --data.init_args.ood_val_path the path to the RoadAnomaly validation dataset
# specify in --data.init_args.img_size the desired input image size
# specify in --data.init_args.oe_enabled to enable Outlier Exposure
# specify in --data.init_args.oe_road_only to paste outlier objects only on road/sidewalk regions
# specify in --data.init_args.oe_perspective_scale to enable perspective scaling of the outlier images
# specify in --data.init_args.ood_prob the probability of inserting an outlier in each training image
# specify in --model.init_args.lr the learning rate for fine-tuning
# specify in --model.init_args.weight_decay the weight decay for fine-tuning
# specify in --model.init_args.inlier_upper_threshold the threshold to consider a pixel as inlier
# specify in --model.init_args.outlier_lower_threshold the threshold to consider a pixel as outlier
# specify in --model.init_args.freeze_encoder to freeze the encoder during fine-tuning
# specify in --model.init_args.num_unfrozen_encoder_blocks the number of unfrozen encoder blocks during fine-tuning
# specify in --model.init_args.outlier_loss_enabled to enable the outlier loss during fine-tuning
# specify in --model.init_args.outlier_weight the weight of the outlier loss during fine-tuning
# specify in --model.init_args.ckpt_path the model checkpoint to fine-tune
!python main.py fit \
  --config configs/dinov2/cityscapes/semantic/eomt_base_640_oe.yaml \
  --trainer.accelerator gpu \
  --trainer.devices 1 \
  --data.init_args.path "/content/drive/MyDrive/EOMT_Project" \
  --data.init_args.coco_root "/content/drive/MyDrive/coco" \
  --data.init_args.coco_split "val" \
  --data.init_args.num_workers 1 \
  --data.init_args.batch_size 1 \
  --data.init_args.ood_val_path "/content/drive/MyDrive/Validation_Dataset/RoadAnomaly" \
  --data.init_args.oe_enabled true \
  --data.init_args.oe_road_only true \
  --data.init_args.oe_perspective_scale false \
  --data.init_args.img_size "[1024, 1024]" \
  --data.init_args.ood_prob 0.2 \
  --model.init_args.lr 1e-4 \
  --model.init_args.weight_decay 0.05 \
  --model.init_args.inlier_upper_threshold -0.6 \
  --model.init_args.outlier_lower_threshold -0.2 \
  --model.init_args.freeze_encoder true \
  --model.init_args.num_unfrozen_encoder_blocks 0 \
  --model.init_args.outlier_loss_enabled true \
  --model.init_args.outlier_weight 1.0 \
  --trainer.accumulate_grad_batches 4 \
  --model.init_args.ckpt_path "/content/drive/MyDrive/EOMT_Project/eomt_cityscapes.bin" \
  --model.init_args.load_ckpt_class_head true \
  --trainer.default_root_dir "/content/drive/MyDrive" \
  --trainer.log_every_n_steps 5
```
